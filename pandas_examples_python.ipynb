```python
# prompt: can you find out basic stats information on my data frame df, calculate measurement such as mean, median, mode, variance, std, range, skewness, kurtosis

# Import necessary libraries if not already imported
import pandas as pd

# Assuming 'df' is already defined and contains your data

# Mean, median, mode, variance, std, range, skewness, kurtosis for the 'salary' column
mean_salary = df['salary'].mean()
median_salary = df['salary'].median()
mode_salary = df['salary'].mode()[0]  # Mode might return multiple values, so we take the first
variance_salary = df['salary'].var()
std_salary = df['salary'].std()
range_salary = df['salary'].max() - df['salary'].min()
skewness_salary = df['salary'].skew()
kurtosis_salary = df['salary'].kurt()


print("Mean:", mean_salary)
print("Median:", median_salary)
print("Mode:", mode_salary)
print("Variance:", variance_salary)
print("Standard Deviation:", std_salary)
print("Range:", range_salary)
print("Skewness:", skewness_salary)
print("Kurtosis:", kurtosis_salary)


# You can repeat these calculations for other relevant numerical columns in your DataFrame.

# Example for the 'service' column:
mean_service = df['service'].mean()
print("Mean Service:", mean_service)
# And so on for other calculations and columns
```
